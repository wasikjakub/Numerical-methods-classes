zwyczajnych. Będziemy rozpatrywać równania różniczkowe postaci
$$\dot{x}(t) = f(x(t), t)$$
gdzie:

$x(t) \in \mathbb{R}^n$,

$t \ge 0$ 


z warunkiem początkowym $x(0) = x0$

Jest to tak zwany problem początkowy (problem Cauchy’ego) dla równań różniczkowych zwyczajnych.

**Zadanie 1.**  
Za pomocą funkcji [solve_ivp](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html#scipy.integrate.solve_ivp) przy wykorzystaniu dwóch metod **RK45** i **RK23** rozwiąż wszystkie równania z poprzedniego laboratorium i porównaj wyniki z zaimplementowaną metodą Eulera w przód.

```python
from scipy import integrate
import matplotlib.pyplot as plt
import numpy as np

def fun(t, y):
    yr = (y + t)/(y - t)
    return yr

t_span = np.linspace(0, 10, 1000)
y0 = np.array([1])

x = sp.integrate.solve_ivp(fun, (0,10), y0, t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metodą RK45')
plt.show()

x = sp.integrate.solve_ivp(fun, (0,10), y0, method='RK23', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metoda RK23')
plt.show()
```

```python
def fun_2(t, y):
    yr = np.zeros((4))
    yr[0] = y[2]
    yr[1] = y[3]
    yr[2] = -(y[0])/(y[0]**2 + y[1]**2)**1.5
    yr[3] = -(y[1])/(y[0]**2 + y[1]**2)**1.5
    return yr 

t_span = np.linspace(0, 10, 1000)
y0 = np.array([1,0,0,1])

x = sp.integrate.solve_ivp(fun_2, (0,10), y0, t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.plot(t_span, x.y[2])
plt.plot(t_span, x.y[3])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metodą RK45')
plt.show()

x = sp.integrate.solve_ivp(fun_2, (0,10), y0, method='RK23', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.plot(t_span, x.y[2])
plt.plot(t_span, x.y[3])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metoda RK23')
plt.show()
```

```python
def fun_3(t, x):
    A = np.array([[-667, 333], [666, -334]])
    B = A @ x
    return B

t_span = np.linspace(0, 10, 1000)
y0 = np.array([0, 3])

x = sp.integrate.solve_ivp(fun_3, (0,10), y0, method='RK45', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metodą RK45')
plt.show()

x = sp.integrate.solve_ivp(fun_3, (0,10), y0, method='RK23', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metoda RK23')
plt.show()
```

```python
def fun_4(t, y):
    mi = 0.012277471
    mi_prim = 1 - mi
    D1 = (((y[0]+mi)**2)+y[2]**2)**1.5 
    D2 = (((y[0]-mi_prim)**2)+y[2]**2)**1.5 
    yr = np.zeros((4))
    yr[0] = y[1]
    yr[1] = y[0]+2*y[3]-mi_prim*(y[0]+mi)/D1-mi*(y[0]-mi_prim)/D2
    yr[2] = y[3]
    yr[3] = y[2]-2*y[1]-mi_prim*y[2]/D1-mi*y[2]/D2
    return yr

t_span = np.linspace(0, 10, 1000)
y0 = np.array([0.994,0,0, -2.00158510637908252240537862224])

x = sp.integrate.solve_ivp(fun_4, (0,10), y0, method='RK45', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.plot(t_span, x.y[2])
plt.plot(t_span, x.y[3])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metodą RK45')
plt.show()

x = sp.integrate.solve_ivp(fun_4, (0,10), y0, method='RK23', t_eval=t_span)

plt.plot(t_span, x.y[0])
plt.plot(t_span, x.y[1])
plt.plot(t_span, x.y[2])
plt.plot(t_span, x.y[3])
plt.grid()
plt.xlabel('Argumenty')
plt.ylabel('Wartości')
plt.title('rozwiazanie numeryczne metoda RK23')
plt.show()
```
