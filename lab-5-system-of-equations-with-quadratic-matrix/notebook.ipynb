```python
m = [10, 20, 50, 100, 1000]
c = []
d = []
for rozmiar in m:
    A, b = main.random_matrix_Ab(rozmiar)
    x = LA.solve(A, b)
    c.append(main.residual_norm(A, x, b))
    d.append(LA.cond(A))

plt.plot(m, c, '-o')
plt.xscale('log')
plt.yscale('log')
plt.title('norma residum w zależności od wielkości macierzy')
plt.show
plt.legend
```

```python

plt.plot(m, d, '-o','r')
plt.xscale('log')
plt.yscale('log')
plt.title('wskaznik uwarunkowania w zależności od wielkosci macierzy')
plt.show
```

Rozkład dowolnej macierzy metodą  *[dekompozycji na wartości singularne](https://pl.wikipedia.org/wiki/Rozk%C5%82ad_wed%C5%82ug_warto%C5%9Bci_osobliwych)* można w Pythonie przeprowadzić przy pomocy funkcji *[svd](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.linalg.svd.html)*. Rozkład dla przykładowej macierzy obrazuje kod: 

```python
import numpy as np
import numpy.linalg as nplin

A = np.array([[1,2,3],[1,2,3],[1,2,3]])

# Użycie rozkładu SVD na macierzy A
U,S,V = nplin.svd(A)

print(S)
# Odtworzenie macierzy A przy pomocy metody SVD
A2 = np.dot(U * S, V)
 
print("Macierz A:\n {0}".format(A))
print("Macierz odtworzona z SVD:\n {0}".format(A2))
```

```python
m = [10, 20, 50, 100, 1000]
wektor_log = []
wektor_max = []
wektor_min = []

for rozmiar in m:
    A, _ = main.random_matrix_Ab(rozmiar)
    sing_log = main.log_sing_value(rozmiar, 0, 100)
    sing_max = main.order_sing_value(rozmiar, site = "gre")
    sing_min = main.order_sing_value(rozmiar, site = "low") 
    A_sing_log = main.create_matrix_from_A(A, sing_log)
    wektor_log.append(np.linalg.cond(A_sing_log))
    A_sing_max = main.create_matrix_from_A(A, sing_max)
    wektor_max.append(np.linalg.cond(A_sing_max))
    A_sing_min = main.create_matrix_from_A(A, sing_min)
    wektor_min.append(np.linalg.cond(A_sing_min))   


plt.yscale('log')
plt.xscale('log')
plt.title('współczynnik uwarunkowania dla wektora w przestrzeni logarytmicznej')
plt.plot(m, wektor_log, '-o')
plt.show() 
```

```python
m = [10, 20, 50, 100, 1000]
residum_log = []
residum_max = []
residum_min = []

for rozmiar in m:
    A, b = main.random_matrix_Ab(rozmiar)
    sing_log = main.log_sing_value(rozmiar, 0, 100)
    sing_max = main.order_sing_value(rozmiar, site = "gre")
    sing_min = main.order_sing_value(rozmiar, site = "low") 
    A_sing_log = main.create_matrix_from_A(A, sing_log)
    x = np.linalg.solve(A_sing_log, b)
    residum_log.append(main.residual_norm(A_sing_log, x, b))
    A_sing_max = main.create_matrix_from_A(A, sing_max)
    x = np.linalg.solve(A_sing_max, b)
    residum_max.append(main.residual_norm(A_sing_max, x, b))
    A_sing_min = main.create_matrix_from_A(A, sing_min)
    x = np.linalg.solve(A_sing_min, b)
    residum_min.append(main.residual_norm(A_sing_min, x, b))
plt.yscale('log')
plt.xscale('log')
plt.title('współczynnik normy residum dla wektora w przestrzeni logarytmicznej')
plt.plot(m, residum_log, '-o')
plt.show() 
```

```python
plt.yscale('log')
plt.xscale('log')
plt.title('współczynnik normy residum dla wektora z jednym duzym elementem')
plt.plot(m, residum_max, '-o')
plt.show()
```

```python
plt.yscale('log')
plt.xscale('log')
plt.title('współczynnik normy residum dla wektora z jednym małym elementem')
plt.plot(m, residum_min, '-o')
plt.show()
```
